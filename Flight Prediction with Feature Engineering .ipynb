```python
#flight_data['Date']=flight_data['Date_of_Journey'].apply[lambda x:x.split('/'[0])]
flight_data['Date']=flight_data['Date_of_Journey'].str.split('/').str[0]
flight_data['Month']=flight_data['Date_of_Journey'].str.split('/').str[1]
flight_data['Year']=flight_data['Date_of_Journey'].str.split('/').str[2]
```

```python
flight_data['Arrival_Hour']=flight_data['Arrival_Time'].str.split(':').str[0]
flight_data['Arrival_Min']=flight_data['Arrival_Time'].str.split(':').str[1]
```

```python
flight_data['Dep_Hour']=flight_data['Dep_Time'].str.split(':').str[0]
flight_data['Dep_min']=flight_data['Dep_Time'].str.split(':').str[1]
flight_data.drop('Dep_Time',axis=1,inplace=True)
```

```python
flight_data['Dep_Hour']=flight_data['Dep_Hour'].astype(dtype=int)
flight_data['Dep_min']=flight_data['Dep_min'].astype(dtype=int)
```

```python
flight_data['Total_Stops']=flight_data['Total_Stops'].map({'non-stop':0,'2 stops':2,'1 stop':1,'3 stops':3,'nan':1,'4 stops':4})
```

```python
flight_data=pd.get_dummies(flight_data,columns=['Airline','Source','Destination','Additional_Info'],drop_first=True)
```

```python
flight_data
```

```python
def hours_mins(duration):
    split_data=duration.split(' ')
    if len(split_data)==2:
        hours=int(split_data[0][:-1])
        mins=int(split_data[1][:-1])
        
    else:
        hours=int(split_data[0][:-1])
        mins=0
    total_mins=hours*60+mins
    return total_mins
```

```python
flight_data['Duration_in_min']=flight_data.Duration.apply(hours_mins)
```

```python
flight_data.drop("Duration",axis=1,inplace=True)
```
